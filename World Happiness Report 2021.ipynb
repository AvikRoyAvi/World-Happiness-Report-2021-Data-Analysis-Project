```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_columns = ['Country name', 'Regional indicator', 'Ladder score', 'Logged GDP per capita', 'Social support',
            'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Ladder score in Dystopia']
```

```python
happy_df = df[data_columns].copy()
```

```python
happy_df.head()
```

```python
happy_df.isnull().sum()
```

```python
#plot between happiness and GDP
plt.figure(figsize = (16, 9))
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
plt.title('Plot Between Happiness and GDP', fontsize = 20)
sns.scatterplot(x = 'Ladder score', y = 'Logged GDP per capita', hue = 'Regional indicator',
                data = happy_df, palette = 'rainbow',s = 200)
plt.xlabel('Happiness Score', fontsize = 15)
plt.ylabel('GDP per capital', fontsize = 15)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
plt.legend()
plt.show()
```

```python
gdp_region = happy_df.groupby('Regional indicator')['Logged GDP per capita'].sum()
gdp_region
```

```python
colors = ['b', 'm', 'r', 'y', 'g', 'c', 'w']
textprops = {"fontsize": 18}
plt.figure(figsize = (20, 10))
gdp_region.plot.pie(autopct = '%1.1f%%', colors = colors, textprops = textprops)
plt.title('GDP by Region', fontsize = 22)
plt.ylabel('',)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
plt.show()
```

```python
#Total countries
total_countries = happy_df.groupby('Regional indicator')['Country name'].count()
print(total_countries)
```

```python
#correlation map
cor = happy_df.corr(method = 'pearson')
plt.figure(figsize = (16, 9))
annot_kws = {'fontsize': 20}
sns.heatmap(cor, mask = np.zeros_like(cor, dtype = np.bool), cmap = 'Blues', square = True, annot_kws = annot_kws)
plt.show()
```

```python
#corruption in region
corruption = happy_df.groupby('Regional indicator')['Perceptions of corruption'].mean()
corruption
```

```python
plt.figure(figsize = (16, 9))
plt.title('Perceptions of Corruption in various Regions', fontsize = 25)
plt.xlabel('Regions', fontsize = 20)
plt.ylabel('Corruption Index', fontsize = 20)
plt.xticks(rotation = 30, ha = 'right', fontsize = 15)
plt.bar(corruption.index, corruption, color = 'r')
plt.show()
```

```python
happy_df = df.rename({'Country name': 'country_name', 'Regional indicator': 'regional_indicator', 
                      'Ladder score': 'Ladder_score', 'Explained by: Healthy life expectancy': 'healthy_life_expectancy', 
                      'Freedom to make life choices': 'Freedom_to_make_life_choices', 'Perceptions of corruption': 'Perceptions_of_corruption'}, axis = 1)
happy_df
```

```python
top_10 = happy_df.head(10)
bottom_10 = happy_df.tail(10)
```

```python
fig, axes = plt.subplots(1, 2, figsize = (16, 6))
plt.tight_layout(pad = 2)
xlabels = top_10.country_name
axes[0].set_title('Top 10 Happiest Countries Life Expectancy', fontsize = 20)
axes[0].set_xticklabels(xlabels, rotation = 45, ha = 'right', fontsize = 12)
sns.barplot(x = top_10.country_name, y = top_10.healthy_life_expectancy, ax = axes[0])
axes[0].set_xlabel('Country Name', fontsize = 15)
axes[0].set_xlabel('Life Expectancy', fontsize = 15)

ylabels = bottom_10.country_name
axes[1].set_title('Top 10 Least Happiest Countries Life Expectancy', fontsize = 20)
axes[1].set_xticklabels(xlabels, rotation = 45, ha = 'right', fontsize = 12)
sns.barplot(x = bottom_10.country_name, y = bottom_10.healthy_life_expectancy, ax = axes[1])
axes[1].set_xlabel('Country Name', fontsize = 15)
axes[1].set_xlabel('Life Expectancy', fontsize = 15)

plt.show()
```

```python
plt.figure(figsize = (16, 9))
sns.scatterplot(x = happy_df.Freedom_to_make_life_choices, y = happy_df.Ladder_score, 
                hue = happy_df.regional_indicator, s= 200)
plt.legend(loc = 'upper left', fontsize = '12')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Happiness score')
plt.show()
```

```python
country = happy_df.sort_values(by = 'Perceptions_of_corruption').head(10)
plt.figure(figsize = (16, 9))
plt.title('Countries with most Perceptions of corruption')
plt.xlabel('Countries', fontsize = 13)
plt.ylabel('Corruption Index', fontsize = 13)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.country_name, country.Perceptions_of_corruption, color = 'y')
plt.show()
```

```python
country = happy_df.sort_values(by = 'Perceptions_of_corruption').tail(10)
plt.figure(figsize = (16, 9))
plt.title('Countries with most Perceptions of corruption')
plt.xlabel('Countries', fontsize = 13)
plt.ylabel('Corruption Index', fontsize = 13)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(country.country_name, country.Perceptions_of_corruption, color = 'g')
plt.show()
```

```python
# corruption vs happiness
plt.figure(figsize = (16, 9))
sns.scatterplot(x = happy_df.Ladder_score, y = happy_df.Perceptions_of_corruption, hue = happy_df.regional_indicator, 
                marker = '*', s = 400)
plt.legend(loc = 'lower left', fontsize = 14)
plt.xlabel('Happiness Score', fontsize = 13)
plt.ylabel('Corruption', fontsize = 13)
plt.show()
```
